```python
def get_top_wells(predictions, target, n_best_wells):
  
    df = pd.DataFrame({
        'predictions': predictions,
        'target': target.reset_index(drop=True)
    })
    
    df_sorted = df.sort_values(by='predictions', ascending=False)
    
    top_wells = df_sorted.head(n_best_wells)
    
    return top_wells
```

```python
def calculate_profit(predictions, target, n_best_wells, budget, revenue_per_1k_barrels,
                     n_points=500, random_state=12345):
  
    df = pd.DataFrame({
        'predictions': predictions,
        'target': target.reset_index(drop=True)
    })
    
    df_sampled = df.sample(n=n_points, random_state=random_state)
    
    df_sorted = df_sampled.sort_values(by='predictions', ascending=False)
    
    top_wells = df_sorted.head(n_best_wells)
    
    total_product = top_wells['target'].sum()
    
    revenue = total_product * revenue_per_1k_barrels
    profit = revenue - budget
    
    return profit, total_product
```

To make a reliable final decision, it is necessary to perform bootstrapping to evaluate risk, as required by the project specification.

```python
def bootstrap_profit(predictions, target, n_best_wells, budget, revenue_per_1k_barrels,
                     n_points=500, n_bootstrap=1000):
    df = pd.DataFrame({
        'predictions': predictions,
        'target': target.reset_index(drop=True)
    })
    
    profits = []
    
    for _ in range(n_bootstrap):
        sample_df = df.sample(n=n_points, replace=True)
        
        top_wells = sample_df.sort_values(by='predictions', ascending=False).head(n_best_wells)
        
        total_product = top_wells['target'].sum()
        revenue = total_product * revenue_per_1k_barrels
        profit = revenue - budget
        
        profits.append(profit)
    
    return profits
```

```python
profits_region_0 = bootstrap_profit(predictions_valid_0, target_valid_0, 
                                    N_BEST_WELLS, BUDGET, REVENUE_PER_1K_BARRELS)

profits_region_1 = bootstrap_profit(predictions_valid_1, target_valid_1, 
                                    N_BEST_WELLS, BUDGET, REVENUE_PER_1K_BARRELS)

profits_region_2 = bootstrap_profit(predictions_valid_2, target_valid_2, 
                                    N_BEST_WELLS, BUDGET, REVENUE_PER_1K_BARRELS)
```

```python
def summarize_bootstrap_results(profits):
    mean_profit = np.mean(profits)
    lower_95 = np.percentile(profits, 2.5)
    upper_95 = np.percentile(profits, 97.5)
    loss_risk = np.mean(np.array(profits) < 0) * 100  
    
    return mean_profit, lower_95, upper_95, loss_risk
```
